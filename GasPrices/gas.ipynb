```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
gas =  pd.read_csv('gas_prices.csv')
plt.figure(figsize=(8,5))
plt.title('Gas prices over time in USD', fontdict={'fontweight':'bold', 'fontsize':22})

""" plt.plot(gas.Year, gas.Germany, 'b.-', label='Germany')
plt.plot(gas.Year, gas.Italy, 'r.-', label='Italy')
plt.plot(gas.Year, gas.France, 'g.-', label='France') """

for country in gas:
    if country != 'Year':
        plt.plot(gas.Year, gas[country], marker='.', label=country)

plt.xticks(gas.Year[::3])
plt.xlabel('Year')
plt.ylabel('US Dollars')
plt.legend(loc='best')# bbox_to_anchor=(1, 0.5))
plt.savefig('GasPrices', dpi=200)
plt.show()
```
